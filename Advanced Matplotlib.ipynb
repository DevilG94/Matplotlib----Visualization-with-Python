```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

## Colored Scatterplots;

```python
iris = pd.read_csv("iris.csv") 
iris.sample(5)
```

```python
# Replacing

iris['Species'] = iris['Species'].replace({'Iris-setosa':0 ,
                        'Iris-versicolor':1 , 'Iris-virginica':2})

iris.sample(5)
```

### C - colors;

```python
plt.scatter( iris['SepalLengthCm'],iris['PetalLengthCm'], c= iris['Species'])    # c

plt.xlabel('Sepal length') 
plt.ylabel('petal length')

plt.show()
```

### Cmap;

```python
# different themes - 'cmap'

plt.scatter( iris['SepalLengthCm'],iris['PetalLengthCm'], c= iris['Species'] ,cmap ='summer') # cmap

plt.xlabel('Sepal length') 
plt.ylabel('petal length')

plt.show()
```

```python
plt.scatter( iris['SepalLengthCm'],iris['PetalLengthCm'], c= iris['Species'] ,cmap ='winter') # winter

plt.xlabel('Sepal length') 
plt.ylabel('petal length')

plt.show()
```

### Color Bar;

```python
plt.scatter( iris['SepalLengthCm'],iris['PetalLengthCm'], c= iris['Species'] ,cmap ='viridis')

plt.xlabel('Sepal length') 
plt.ylabel('petal length')

plt.colorbar()        # colorbar shows variations levels

plt.show()
```

### Alpha;

```python
plt.scatter( iris['SepalLengthCm'],iris['PetalLengthCm'], c= iris['Species'] ,cmap ='jet',
                                alpha =0.5) # reduces,increase color

plt.xlabel('Sepal length') 
plt.ylabel('petal length')
plt.colorbar() 
plt.show()
```

### Plot size;

```python
plt.figure(figsize=(15,7)) # 15 -width , 7 -height

plt.scatter( iris['SepalLengthCm'],iris['PetalLengthCm'], c= iris['Species'] ,
            cmap ='jet', alpha =0.5)	# reduces color

plt.xlabel('Sepal length') 
plt.ylabel('petal length')
plt.colorbar() 
plt.show()
```

```python
# on Data

fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))

# axis
ax[0].scatter(batter['avg'],batter['strike_rate'],color='red') 
ax[1].scatter(batter['avg'],batter['runs'],color ='green')

ax[0].set_title('Avg Vs Strike Rate') 
ax[0].set_ylabel('Strike Rate')

ax[1].set_title('Avg Vs Runs') 
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

# its 2D array we have to metion (0,0)(0,1)(1,0)(1,1)

ax[0,0].scatter(batter['avg'],batter['strike_rate'],color='red') # gives first set of axis

ax[0,1].scatter(batter['avg'],batter['runs'],color='gold') # gives	second set of axis

ax[1,0].hist(batter['avg'],color='cyan') # gives third set of axis 

ax[1,1].hist(batter['runs'],color='pink') # gives fourth set of axis 

fig.show()
```

```python
# Manual way of adding subplots

fig = plt.figure(figsize=(10,10))

ax1 = fig.add_subplot(2,2,1) #2 rows , 2 column , 1 graph
ax1.scatter(batter['avg'],batter['strike_rate'],color='red')

ax2 = fig.add_subplot(2,2,2) #2 rows , 2 column , 2 graph
ax2.hist(batter['runs'],color='gold')

ax3 = fig.add_subplot(2,2,3) #2 rows , 2 column , 3 graph
ax3.hist(batter['avg'],color ='green')

fig.show()
```

# 3D Surface Plot;

```python
# Loss function x2 + y2

# Helpful in Machine learning

x = np.linspace(-10,10,100) 
y = np.linspace(-10,10,100)
```

```python
x
```

```python
y
```

```python
np.meshgrid(x,y)
```

```python
# Mesh grid

xx,yy = np.meshgrid(x,y)
```

```python
xx.shape
```

```python
yy.shape
```

```python
z = xx**2 + yy**2
```

```python
z.shape
```

### 3D Surface Plot;

```python
fig =plt.figure(figsize =(12,8))

ax = plt.subplot(projection = '3d')

ax.plot_surface(xx,yy,z, color ='blue')
```

```python
# Adding theme

fig = plt.figure(figsize=(12,8)) 
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis') 
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8)) 
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='winter')

fig.colorbar(p)
```

```python
z = np.sin(xx) + np.sin(yy)

fig = plt.figure(figsize=(12,8)) 
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis') 
fig.colorbar(p)
```

## Contour Plots;

#### Representing 3D to 2D;

```python
#3D graph
fig = plt.figure(figsize=(12,8)) 
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis') 
fig.colorbar(p)
```

```python
# Contour Plot

fig = plt.figure(figsize=(12,8)) 
ax = plt.subplot()
p = ax.contour(xx,yy,z,cmap='viridis') 
fig.colorbar(p)
```

## Contourf Plot;

```python
fig = plt.figure(figsize=(12,8)) 
ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap='viridis') 
fig.colorbar(p)

# Here blue colour is bend/shallow and yellow is up/top in 3D
```

```python
z = np.sin(xx) + np.sin(yy)

fig = plt.figure(figsize=(12,8)) 
ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap='viridis') 
fig.colorbar(p)
```

```python
# Can be used on a dataframe as well

import seaborn as sns
```

```python
tips = sns.load_dataset('tips') 
tips.head()
```

```python
tips['size'] = tips['size'] * 100
```

```python
tips.head()
```

```python
# Scatter plot -> labels -> markers -> figsize -> color -> cmap

tips.plot(kind='scatter',x='total_bill',y='tip', title='Cost Analysis',marker='+',figsize=(10,6),
          s='size',c='sex', cmap='viridis')
```

```python
tips.groupby('sex')['total_bill'].mean().plot(kind='bar', color ='pink')
```

### On Multiindex Dataframes:-

```python
# Using tips

tips.pivot_table(index=['day','time'],
columns=['sex','smoker'], values='total_bill',
aggfunc='mean').plot(kind='pie',
subplots=True,
figsize=(30,20))
plt.show()
```
