## Histogram;

```python
# using stocks

stocks[['MSFT','FB']].plot(kind='hist',bins=40 , colormap ='summer')
```

### Single and Multiple:-

```python
df = pd.DataFrame(
{
'batsman':['Dhawan','Rohit','Kohli','SKY','Pandya','Pant'], 'match1':[120,90,35,45,12,10],
'match2':[0,1,123,130,34,45],
'match3':[50,24,145,45,10,90]
}
)
```

```python
df.head()
```

```python
df['match1'].plot(kind='pie',labels=df['batsman'].values,autopct='%0.1f%%', colormap= 'tab10')
```

```python
# Multiple pie charts

df[['match1','match2','match3']].plot(kind='pie',subplots=True, figsize=(15,8),colormap='Paired')
```
